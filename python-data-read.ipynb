## Multifile Plotting

```python
df1 = pd.read_csv('data/G-01.txt', delimiter='\t', header=None)
df2 = pd.read_csv('data/G-02.txt', delimiter='\t', header=None)
df3 = pd.read_csv('data/G-03.txt', delimiter='\t', header=None)

df1 = df1.rename(columns={0: 'Frequency', 1: 'Transmittance'})
df2 = df2.rename(columns={0: 'Frequency', 1: 'Transmittance'})
df3 = df3.rename(columns={0: 'Frequency', 1: 'Transmittance'})
```

```python
sns.lineplot(data=df1, x='Frequency', y='Transmittance')
sns.lineplot(data=df2, x='Frequency', y='Transmittance')
sns.lineplot(data=df3, x='Frequency', y='Transmittance')
```

```python
df = pd.DataFrame()

df['Frequency'] = df1['Frequency']
df['Transmittance (Graphine)'] = df1['Transmittance']
df['Transmittance (Triazine)'] = df2['Transmittance']
df['Transmittance (Complex)'] = df3['Transmittance']
```

```python
df
```

```python
sns.lineplot(data=df, x='Frequency', y='Transmittance (Graphine)')
sns.lineplot(data=df, x='Frequency', y='Transmittance (Triazine)')
sns.lineplot(data=df, x='Frequency', y='Transmittance (Complex)')
```

## Saving DataFrames as CSV

```python
df.to_csv('data/merged.csv', header=False)
```
